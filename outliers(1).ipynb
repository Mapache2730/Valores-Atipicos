```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el archivo limpio
df = pd.read_csv("MexicoLimpio.csv", encoding="ISO-8859-1")

# Contar valores nulos en cada columna
valores_nulos = df.isnull().sum()
print("Valores nulos por columna:\n", valores_nulos)

# Seleccionar variables cuantitativas (columnas numricas)
cuantitativas = df.select_dtypes(include=[np.number])

# 1. Mtodo Desviacin Estndar para eliminar outliers
def eliminar_outliers_desviacion(df, columna, z_score_threshold=3):
    media = df[columna].mean()  # Media de la columna
    desviacion = df[columna].std()  # Desviacin estndar de la columna
    z_scores = (df[columna] - media) / desviacion  # Clculo manual
    df_sin_outliers = df[abs(z_scores) < z_score_threshold]  # Filtrar los outliers
    return df_sin_outliers

# 2. Mtodo Rango Intercuartlico (IQR) para eliminar outliers
def eliminar_outliers_iqr(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_sin_outliers = df[(df[columna] >= lower_bound) & (df[columna] <= upper_bound)]
    return df_sin_outliers

# Eliminar outliers de cada columna cuantitativa y mostrar boxplots
for columna in cuantitativas.columns:
    # Mtodo Desviacin Estndar
    df_sin_outliers_desviacion = eliminar_outliers_desviacion(df, columna)
    
    # Mtodo Rango Intercuartlico (IQR)
    df_sin_outliers_iqr = eliminar_outliers_iqr(df, columna)
    
    # Grfico de caja para cada mtodo
    fig, axes = plt.subplots(1, 2, figsize=(15, 8))
    axes[0].boxplot(df[columna], vert=False)
    axes[0].set_title(f"Boxplot original de {columna}")
    axes[1].boxplot(df_sin_outliers_desviacion[columna], vert=False)
    axes[1].set_title(f"Boxplot sin outliers (Desviacin Estndar) de {columna}")
    plt.show()

    # Grfico de caja para el mtodo IQR
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.boxplot(df_sin_outliers_iqr[columna], vert=False)
    ax.set_title(f"Boxplot sin outliers (IQR) de {columna}")
    plt.show()

    print(cuantitativas.columns)

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el archivo limpio
df = pd.read_csv("MexicoLimpio.csv", encoding="ISO-8859-1")

# Contar valores nulos en cada columna
valores_nulos = df.isnull().sum()
print("Valores nulos por columna:\n", valores_nulos)

# Seleccionar variables cualitativas (columnas categricas)
cualitativas = df.select_dtypes(include=[object])

# Funcin para detectar outliers en variables cualitativas
def eliminar_outliers_cualitativas(df, columna, umbral=0.01):
    # Contar la frecuencia de cada categora
    frecuencias = df[columna].value_counts(normalize=True)  # Proporcin de cada categora
    print(f"Frecuencia de categoras en {columna}:\n", frecuencias)
    
    # Identificar las categoras con frecuencia menor que el umbral
    categorias_outliers = frecuencias[frecuencias < umbral].index
    print(f"Categoras con frecuencia menor que {umbral}:\n", categorias_outliers)
    
    # Reemplazar categoras raras con 'Otro'
    df[columna] = df[columna].replace(categorias_outliers, 'Otro')
    return df

# Aplicar la funcin de eliminacin de outliers en las columnas cualitativas
for columna in cualitativas.columns:
    df = eliminar_outliers_cualitativas(df, columna)
    
    # Convertir las categoras a valores numricos para representarlas en boxplot
    df['categoria_num'] = pd.Categorical(df[columna]).codes  # Asigna cdigos numricos a las categoras
    
    # Graficar el boxplot para la columna cualitativa representada como numrica
    plt.figure(figsize=(10, 6))
    plt.boxplot(df['categoria_num'], vert=False)
    plt.title(f"Boxplot para {columna} (con outliers tratados)")
    plt.xlabel("Categoras")
    plt.yticks([1], [columna])  # Etiquetamos el eje Y para que sea el nombre de la columna
    plt.show()

# Mostrar los cambios realizados
print("Dataframe despus de tratar outliers cualitativos:\n", df.head())

```
